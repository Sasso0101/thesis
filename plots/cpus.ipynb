```python
import matplotlib.pyplot as plt

import pandas as pd

data = pd.DataFrame({
  'Year': list(range(2013, 2025+1)),
  'AMD': [16, 16, 16, 16, 32, 32, 64, 64, 64, 96, 128, 192, 192],
  'Intel': [12, 18, 18, 22, 28, 28, 56, 56, 56, 56, 64, 144, 144]
})

data = data.melt(['Year'], 
             var_name='Type', 
             value_name='Count')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

with sns.plotting_context("talk"):
  # Create a figure and axes object for more control over the plot
  fig, ax = plt.subplots(figsize=(12, 6))

  palette = {
    'AMD': '#ED1C24',   # Red color for AMD
    'Intel': '#0071C5'  # Blue color for Intel
  }

  g = sns.barplot(
    data=data,
    x='Year',
    y='Count',
    hue='Type',
    linewidth=1,
    ax=ax,
    palette=palette,
    zorder=2
  )

  ax.set_xlabel("Year")
  ax.set_ylabel("Core count (per socket)")
  ax.legend()
  plt.grid(axis='y')
  fig.patch.set_alpha(0.0)
  ax.patch.set_alpha(0.0)
  
  plt.savefig('cores.png', 
    dpi=300, 
    bbox_inches='tight', 
    facecolor='white', 
    edgecolor='none',
    format='png')

  plt.show()
```
